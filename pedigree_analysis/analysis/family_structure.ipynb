```python
def calc_trios(pedigree):
    trios_child = []
    for id in pedigree.loc[pedigree["id"].notna(), "id"]:        
            # with open("trios.txt") as trio_file:
            mom = pedigree.loc[pedigree["id"] == id, "dam_id"].values[0]
            dad = pedigree.loc[pedigree["id"] == id, "sire_id"].values[0]
            if mom in pedigree["id"].values and dad in pedigree["id"].values:
                if pd.notna(pedigree.loc[pedigree["id"] == mom, "id_bed"].values[0]) and pd.notna(pedigree.loc[pedigree["id"] == dad, "id_bed"].values[0]):
                     trios_child.append(id)
    return trios_child
```

```python
def calc_siblings(pedigree):
    full_siblings = {}  # Initialize as a dictionary
    half_sibling_m = {}
    half_sibling_d = {}
    
    # Iterate through each individual
    for idx, row in pedigree.iterrows():
        id_ = row["id"]
        mom = row["dam_id"]
        dad = row["sire_id"]
        
        # Skip if individual ID is NaN
        if pd.isna(id_):
            continue
        
        # Get siblings through mother
        if not pd.isna(mom):
            if mom in half_sibling_m:
                half_sibling_m[mom].append(id_)
            else:
                half_sibling_m[mom] = [id_]
        
        # Get siblings through father
        if not pd.isna(dad):
            if dad in half_sibling_d:
                half_sibling_d[dad].append(id_)
            else:
                half_sibling_d[dad] = [id_]
        
        # Check for full siblings
        if not pd.isna(mom) and not pd.isna(dad):
            parents = (mom, dad)
            if parents in full_siblings:
                full_siblings[parents].append(id_)
            else:
                full_siblings[parents] = [id_]
    
    half_sibling_m = {mom: children for mom, children in half_sibling_m.items() if len(children) > 1}
    half_sibling_d = {dad: children for dad, children in half_sibling_d.items() if len(children)>1}
    full_siblings = {parents: children for parents, children in full_siblings.items() if len(children) > 1}
    
    return half_sibling_m, half_sibling_d, full_siblings
```

```python
half_sibling_m, half_sibling_d, full_siblings = calc_siblings(ped_df)
```

```python
def calc_half_sibling_counts(pedigree):
    # Dictionaries to store sibling counts by their parents
    maternal_siblings = {}
    paternal_siblings = {}
    
    # Iterate through each individual in the pedigree dataset
    for idx, row in pedigree.iterrows():
        id_ = row["id"]
        mom = row["dam_id"]
        dad = row["sire_id"]
        
        # Skip if individual ID is NaN
        if pd.isna(id_):
            continue
        
        # Collect maternal siblings
        if not pd.isna(mom):
            if mom not in maternal_siblings:
                maternal_siblings[mom] = []
            maternal_siblings[mom].append((id_, dad))
        
        # Collect paternal siblings
        if not pd.isna(dad):
            if dad not in paternal_siblings:
                paternal_siblings[dad] = []
            paternal_siblings[dad].append((id_, mom))
    
    # Calculate half-sibling counts
    half_sibling_counts_m = {}
    half_sibling_counts_d = {}
    
    for mom, children in maternal_siblings.items():
        if len(children) > 1:
            unique_fathers = set(dad for _, dad in children)
            if len(unique_fathers) > 1:
                half_sibling_counts_m[mom] = len(children)
    
    for dad, children in paternal_siblings.items():
        if len(children) > 1:
            unique_mothers = set(mom for _, mom in children)
            if len(unique_mothers) > 1:
                half_sibling_counts_d[dad] = len(children)
    
    return half_sibling_counts_m, half_sibling_counts_d
```

```python
half_sibling_counts_m, half_sibling_counts_d = calc_half_sibling_counts(ped_df)
```

```python
def calculate_generations(pedigree_df):
    # Find the oldest and youngest individuals
    oldest_yob = pedigree_df['YOB'].min()
    youngest_yob = pedigree_df['YOB'].max()

    # Calculate age differences between parents and children
    age_diffs = []
    for _, row in pedigree_df.iterrows():
        if pd.notna(row['YOB']):
            if pd.notna(row['sire_id']):
                sire_row = pedigree_df.loc[pedigree_df['id'] == row['sire_id']]
                if not sire_row.empty:
                    sire_yob = sire_row.iloc[0]['YOB']
                    age_diff_sire = row['YOB'] - sire_yob
                    age_diffs.append(age_diff_sire)
            if pd.notna(row['dam_id']):
                dam_row = pedigree_df.loc[pedigree_df['id'] == row['dam_id']]
                if not dam_row.empty:
                    dam_yob = dam_row.iloc[0]['YOB']
                    age_diff_dam = row['YOB'] - dam_yob
                    age_diffs.append(age_diff_dam)

    # Calculate the number of generations
    if age_diffs:
        average_generation_age_diff = sum(age_diffs) / len(age_diffs)
        number_of_generations = (youngest_yob - oldest_yob) / average_generation_age_diff
    else:
        number_of_generations = 0
    
    return average_generation_age_diff, number_of_generations

 
average_generation_age_diff, number_of_generations = calculate_generations(ped_df)
print("Average generation age difference:", average_generation_age_diff,  
      "Number of generations in the pedigree:", number_of_generations)
```
